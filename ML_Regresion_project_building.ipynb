```python
#outliners Handling
#iqr
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3-q1
lower_limit = q1-1.5*iqr
upper_limit = q3+1.5*iqr
print(lower_limit)
print(upper_limit)
```

```python
#exclude the data points
```

```python
df[(df<lower_limit) | (df>upper_limit)]
```

```python
df[~((df<lower_limit) | (df>upper_limit))]  # ~ symbol is exclude
```

```python
df = df[~((df<lower_limit)|(df>upper_limit)).any(axis=1)]
df
```

```python
x = df.drop('SepalLengthCm',axis=1)
y =df['SepalLengthCm']
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
x_test.shape,y_test.shape
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
#Lasso and Ridge -Regulation
#Whenever over the issue of overfitting
```

```python
models ={
    'LinearRegression':LinearRegression(),
    'LassoRegression':Lasso(alpha=0.03),
    'RidgeRegrassion':Ridge(alpha=20)
}
```

```python
result = {}
for name,model in models.items():
  print(name,model)
```

```python
result = {}
for name,model in models.items():
  model.fit(x_train,y_train)
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test,y_pred)
  mae = mean_absolute_error(y_test,y_pred)
  r2 = r2_score(y_test,y_pred)
  result[name] = {'mse':mse,'mae':mae,'r2':r2}
```
