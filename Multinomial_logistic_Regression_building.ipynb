```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report,roc_auc_score
from sklearn.datasets import load_iris
```

```python
import sklearn.datasets as datasets
```

```python
data = load_iris()
data
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target_names
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df
```

```python
df['target'] = data.target
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
#split the data
```

```python
x =df.drop(columns=['target'])
y = df['target']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#train multinomial logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(multi_class='multinomial',solver='lbfgs')
model
```

```python
#model training
```

```python
model.fit(x_train,y_train)
```

```python
#prediction
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
acc =accuracy_score(y_test,y_pred)
acc
```

```python
precision = precision_score(y_test,y_pred,average='weighted')
precision
```

```python
recall = recall_score(y_test,y_pred,average='weighted')
recall
```

```python
f1=f1_score(y_test,y_pred,average='weighted')
f1
```

```python
print(classification_report(y_test,y_pred))
```

```python
#L1 and L2 regulation for better generalization
```

```python
model_l2=LogisticRegression(multi_class='multinomial',solver='lbfgs',penalty='l2',C=1.0)
model_l2
```

```python
model_l2.fit(x_train,y_train)
```

```python
y_pred_l2=model_l2.predict(x_test)
y_pred_l2
```

```python
print(classification_report(y_test,y_pred_l2))
```

```python
#Hyperparameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    'C':[0.01,0.1,1,10,100],
    'penalty':['l1','l2'],  #ridge regulation, l1 not support for multinomial in lbfgs
    'solver':['lbfgs']

}
```

```python
param_grid
```

```python
from math import log
log_reg =LogisticRegression(multi_class='multinomial',max_iter=200)
log_reg
```

```python
grid_search=GridSearchCV(log_reg,param_grid,cv=5,scoring='accuracy',n_jobs=1)
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
y_pred_best=best_model.predict(x_test)
y_pred_best
```

```python
print(classification_report(y_test,y_pred_best))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search=RandomizedSearchCV(log_reg,param_grid,cv=5,random_state=10,n_iter=10)
random_search
```

```python
random_search.fit(x_train,y_train)
```

```python
y_pred_best2=random_search.predict(x_test)
y_pred_best2
```

```python
print(classification_report(y_test,y_pred_best2))
```
