```python
#missing value Handling
```

```python
import pandas as pd
import numpy as np
```

```python
data = {"Name":['ajay','suresh','david','pragg','gukesh'],
        "Age":[25,np.nan,30,np.nan,20],
        "Salary":[50000,60000,np.nan,70000,80000],
        "City":['new york','los angles','new york',np.nan,'chocago']}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#missing value heatmap
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.isnull())
plt.title("missing value")
plt.show()
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)  #removes column
```

```python
#filling missing value
# 1 method
#filling with constsant
```

```python
df.fillna({'Age':0,'Salary':0,'City':'Unknown'})
```

```python
df_new =df.fillna({'Age':0,'Salary':0,'City':'Unknown'})
df_new
```

```python
#method 2
#handling missing value using mean
# df['Age'] = df['Age'].fillna(mean_age)
# df['Age'].fillna(df['age].mean(),inplace=True)
```

```python
print(df['Age'].mean())
```

```python
mean_age = df['Age'].mean()
```

```python
df['Age'].fillna(mean_age)
```

```python
df['Age'] = df['Age'].fillna(mean_age)
df
```

```python
df['Salary']
```

```python
df['Salary'].mean()
```

```python
df['Salary'].median()
```

```python
median_salary = df['Salary'].median()
```

```python
df['Salary'].fillna(median_salary,inplace=True)
df
```

```python
# incase of categorical column to handle missing value - mode(method)
```

```python
df['City']
```

```python
df['City'].value_counts()
```

```python
df['City'].mode()
```

```python
df['City'].mode()[0]
```

```python
city_mode = df['City'].mode()[0]
```

```python
df['City'].fillna(city_mode)
```

```python
df
```

```python
df['City'].fillna(city_mode,inplace=True)
df
```
