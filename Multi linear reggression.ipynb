```python
#scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(df[['SquareFootage','Price']])
```

```python
df.head(5)
```

```python
df[['SquareFootage','Price']] = scaler.fit_transform(df[['SquareFootage','Price']])
df.head(5)
```

```python
#define dependent and independent features
```

```python
df.columns
```

```python
x = df[['SquareFootage', 'Price', 'bedroom ', 'locaation ']]
y = df['Price']
```

```python
#spliting the data into training and test
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
#building linear regression mpdel
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
#train the model using fit method
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Make prediction
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#model intercept and model slope
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
MSE = mean_squared_error(y_test,y_pred)
MSE
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MAE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
R2 = r2_score(y_test,y_pred)
R2
```
