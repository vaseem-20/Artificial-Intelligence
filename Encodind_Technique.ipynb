```python
fig,axes=plt.subplots(1,2,figsize=(8,4))
sns.histplot(df['sqrt_PetalLengthCm'],ax=axes[0],kde=True)
sns.histplot(df['sqrt_PetalWidthCm'],ax=axes[1],kde=True)
```

```python
# Power Transformation or Box - Cox Trnsformation
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
power_transformer = PowerTransformer(method='box-cox')
power_transformer
```

```python
power_transformer.fit_transform(df[['PetalLengthCm','PetalWidthCm']])
```

```python
df[['boxcox_PetalLengthCm','boxcox_PetalWidthCm']]=power_transformer.fit_transform(df[['PetalLengthCm','PetalWidthCm']])
df[['boxcox_PetalLengthCm','boxcox_PetalWidthCm']]
```

```python
df
```

```python
fig,axes=plt.subplots(1,2,figsize=(8,4))
sns.histplot(df['boxcox_PetalLengthCm'],ax=axes[0],kde=True)
sns.histplot(df['boxcox_PetalWidthCm'],ax=axes[1],kde=True)
```
