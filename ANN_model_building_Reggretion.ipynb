```python
# split the data
```

```python
x=df.drop('ee_consumption',axis=1).values
y=df['ee_consumption'].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler
```

```python
x_train_scale=scaler.fit_transform(x_train)
```

```python
x_test_scaler=scaler.transform(x_test)
x_test_scaler
```

```python
x_train.shape
```

```python
# model architecture
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
model=Sequential()
model
```

```python
model.add(tf.keras.layers.Input(shape=[4],name='input_layer'))
model.add(Dense(units=64,activation='relu',name='hidden_layer_1'))
model.add(Dense(units=32,activation='relu',name='hidden_layer_2'))
model.add(Dense(units=16,activation='relu',name='hidden_layer_3'))
model.add(Dense(units=1,name='output_layer'))
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error',

  )
```

```python
model.fit(
    x=x_train_scale,
    y=y_train,
    epochs=100,
    batch_size=32,
    validation_data=(x_test_scaler,y_test),
    callbacks=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
)
```

```python
pd.DataFrame(model.history.history)
```

```python
model.evaluate(x_test_scaler,y_test)
```

```python
model.evaluate(x_train_scale,y_train)
```

```python
prediction = model.predict(x_test_scaler)
prediction
```

```python
prediction_df=pd.DataFrame(np.ravel(prediction),columns=['prediction'])
prediction_df
```

```python
compare_df=pd.concat([pd.DataFrame(y_test,columns=['true_values']),prediction_df],axis=1)
compare_df
```
